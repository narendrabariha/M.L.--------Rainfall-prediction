```python
X = df.drop(['PrecipitationSumInches'], axis = 1)
```

```python
Y = df["PrecipitationSumInches"]
```

```python
#reshaping it into 2d vector 
Y = Y.values.reshape(-1,1)
```

```python
day_index = 798
days = [ i for i in range(Y.size)]
```

```python
#initialise the linear regression classifier 
clf=LinearRegression()

#train the classifier 
clf.fit(X,Y)
```

```python
#plot a graph
print("The precipitation trend graph:")
plt.scatter (days, Y, color = 'g')
plt.scatter(days [day_index], Y[day_index], color= 'r')
plt.title("Precipitation level")
plt.xlabel("Days")
plt.ylabel("Precipitation in inches")
plt.show()
x_vis = X.filter(['TempAvgF', 'DewPointAvgF', 'HumidityAvgPercent', 'SeaLevelPressureAvgInches', 'VisibilityAvgMiles', 'WindAvgMPH'])
```

```python
print("The precipitation vs attributes trend graph:")

for i in range(x_vis.columns.size):
    plt.subplot(3,2,i+1)
    plt.scatter (days, x_vis[x_vis.columns.values[i][:100]], color = 'g') 
    plt.scatter(days [day_index],x_vis[x_vis.columns.values[i]][day_index], color = 'r')  
    plt.title(x_vis.columns.values[i])
    plt.tight_layout() 
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.model_selection import train_test_split

# Assuming X and Y are your features and target variables
# Split the data into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Initialize the Linear Regression model
clf = LinearRegression()

# Train the model
clf.fit(X_train, Y_train)

# Predict on training and test sets
Y_train_pred = clf.predict(X_train)
Y_test_pred = clf.predict(X_test)

# Evaluate the model
train_mse = mean_squared_error(Y_train, Y_train_pred)
test_mse = mean_squared_error(Y_test, Y_test_pred)

train_r2 = r2_score(Y_train, Y_train_pred)
test_r2 = r2_score(Y_test, Y_test_pred)

train_mae = mean_absolute_error(Y_train, Y_train_pred)
test_mae = mean_absolute_error(Y_test, Y_test_pred)

print("Training Set Metrics:")
print(f"Mean Squared Error (MSE): {train_mse:.2f}")
print(f"R-squared (R2): {train_r2:.2f}")
print(f"Mean Absolute Error (MAE): {train_mae:.2f}")
print()
print("Test Set Metrics:")
print(f"Mean Squared Error (MSE): {test_mse:.2f}")
print(f"R-squared (R2): {test_r2:.2f}")
print(f"Mean Absolute Error (MAE): {test_mae:.2f}")
```

```python
clf.score(X_train,Y_train)
```

```python
clf.score(X_test,Y_test)
```

```python
from sklearn.linear_model import Ridge, Lasso

# Example of Ridge Regression
ridge = Ridge(alpha=0.1)  # Adjust alpha for regularization strength
ridge.fit(X_train, Y_train)
print("Ridge Regression Train Score:", ridge.score(X_train, Y_train))
print("Ridge Regression Test Score:", ridge.score(X_test, Y_test))

# Example of Lasso Regression
lasso = Lasso(alpha=0.1)  # Adjust alpha for regularization strength
lasso.fit(X_train, Y_train)
print("Lasso Regression Train Score:", lasso.score(X_train, Y_train))
print("Lasso Regression Test Score:", lasso.score(X_test, Y_test))
```

```python
from sklearn.model_selection import cross_val_score

# Example of cross-validation with Linear Regression
scores = cross_val_score(clf, X, Y, cv=5)  # Adjust cv for number of folds
print("Cross-validation Scores:", scores)
```
